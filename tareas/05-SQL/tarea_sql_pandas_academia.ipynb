## 3. Cargar tablas a SQLite

```python
conn = sqlite3.connect(":memory:")

estudiantes.to_sql("estudiantes", conn, index=False, if_exists="replace")
cursos.to_sql("cursos", conn, index=False, if_exists="replace")
inscripciones.to_sql("inscripciones", conn, index=False, if_exists="replace")
pagos.to_sql("pagos", conn, index=False, if_exists="replace")

def run_sql(query):
    return pd.read_sql_query(query, conn)
```

### Punto 2. Ordenamiento

Trae los cursos ordenados de **mayor a menor precio** y muestra solo:

### Punto 3. Filtro con múltiples condiciones

### Punto 4. Agregación simple

Calcula:
- número total de cursos
- precio promedio
- precio mínimo
- precio máximo

Usa la tabla `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 4 - SQL
# Escribe aquí tu consulta

query = """
SELECT COUNT(*) AS total_cursos, AVG(precio) AS precio_promedio, MAX(precio) AS precio_maximo, MIN(precio) AS precio_minimo
FROM cursos 
"""
resultado_sql4 = run_sql(query)
resultado_sql4
```

```python
# Punto 4 - pandas
# Escribe aquí tu solución

resultado_pd4 = pd.DataFrame({
    "total_cursos": [cursos.shape[0]],
    "precio_promedio": [cursos["precio"].mean()],
    "precio_maximo": [cursos["precio"].max()],
    "precio_minimo": [cursos["precio"].min()]
})

resultado_pd4
```

### Punto 5. GROUP BY

Cuenta cuántos cursos hay por `categoria`.

Ordena de mayor a menor cantidad.

**Entregable:**
- SQL
- pandas

```python
# Punto 5 - SQL
# Escribe aquí tu consulta

query = """
SELECT categoria, COUNT(*) AS total_cursos
FROM cursos
GROUP BY categoria
ORDER BY total_cursos DESC
"""

resultado_sql5 =run_sql(query)
resultado_sql5
```

```python
# Punto 5 - pandas
# Escribe aquí tu solución

resultado_pd5 = cursos.groupby("categoria").size().reset_index(name = "total_cursos").sort_values(by = "total_cursos", ascending = False)
resultado_pd5
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
# Escribe aquí tu consulta

query = """
SELECT e.nombre AS nombre_estudiante, c.nombre_curso, i.estado
FROM inscripciones i
JOIN estudiantes e ON i.estudiante_id = e.estudiante_id
JOIN cursos c ON i.curso_id = c.curso_id

"""
resultado_sql6 = run_sql(query)
resultado_sql6
```

```python
# Punto 6 - pandas
# Escribe aquí tu solución

resultado_pd6 = inscripciones.merge(estudiantes, on="estudiante_id").merge(cursos, on="curso_id")[["nombre", "nombre_curso", "estado"]]
resultado_pd6
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
# Escribe aquí tu consulta

query = """
SELECT e.estudiante_id, e.nombre, i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
"""

resultado_sql7 = run_sql(query)
resultado_sql7
```

```python
# Punto 7 - pandas
# Escribe aquí tu solución

resultado_pd7 = estudiantes.merge(inscripciones, on="estudiante_id", how="left")[["estudiante_id", "nombre", "inscripcion_id"]]
resultado_pd7
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
# Escribe aquí tu consulta

query = """
SELECT e.nombre, COUNT(i.inscripcion_id) AS total_inscripciones
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
GROUP BY e.nombre
HAVING total_inscripciones >= 2
"""

resultado_sql8 = run_sql(query)
resultado_sql8
```

```python
# Punto 8 - pandas
# Escribe aquí tu solución

resultado_pd8 = estudiantes.merge(inscripciones, on="estudiante_id").groupby("nombre").size().reset_index(name = "total_inscripciones")
resultado_pd8 = resultado_pd8[resultado_pd8["total_inscripciones"] >= 2]
resultado_pd8
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
# Escribe aquí tu consulta

query = """
SELECT nombre_curso, precio,
CASE 
    WHEN precio < 120 THEN "bajo"
    WHEN precio BETWEEN 120 AND 180 THEN "medio"
    WHEN precio > 180 THEN "alto"
    END AS segmento_precio
    FROM cursos
"""
resultado_sql9 = run_sql(query)
resultado_sql9
```

```python
# Punto 9 - pandas
# Escribe aquí tu solución

resultado_pd9 = cursos[["nombre_curso", "precio"]].copy()
resultado_pd9["segmento_precio"] = pd.cut(resultado_pd9["precio"], bins=[0, 120, 180, float("inf")], labels = ["bajo", "medio", "alto"])
resultado_pd9
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
# Escribe aquí tu consulta

query = """
SELECT c.categoria, SUM(p.monto) AS total_pagado
FROM cursos c
JOIN inscripciones i ON c.curso_id = i.curso_id
JOIN pagos p ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC
"""
resultado_sql10 = run_sql(query)
resultado_sql10
```

```python
# Punto 10 - pandas
# Escribe aquí tu solución

resultado_pd10 = cursos.merge(inscripciones, on="curso_id").merge(pagos, on="inscripcion_id").groupby("categoria")["monto"].sum().reset_index(name = "total_pagado").sort_values(by = "total_pagado", ascending = False)
resultado_pd10
```
