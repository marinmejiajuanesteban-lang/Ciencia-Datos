**¿Cual es el nombre de la persona con mejor pago (incluyendo beneficios)?**

```python
# Copie el código aca
#Primero debemos encontarar el valor max de totalpaybenefits
max_totalpaybenefits = salarios["TotalPayBenefits"].max()
print(f"El valor maximo pagado es: {max_totalpaybenefits:.2f}")

#Averiguamos cuantas veces se repite ese valor maximo
cantidad_max_totalpaybenefits = (salarios["TotalPayBenefits"] == max_totalpaybenefits).sum()
print(f"El valor maximo de TotalPayBenefits se repite {cantidad_max_totalpaybenefits} veces")

#Miramos quien es la persona que se gana ese dinero
print(f"El empleado que mas gana es: {salarios.loc[salarios["TotalPayBenefits"] == max_totalpaybenefits, "EmployeeName"].values[0]}")
```

```python
# Copie el código aca
#Primero debemos encontarar el valor min de totalpaybenefits
min_totalpaybenefits = salarios["TotalPayBenefits"].min()
print(f"El valor minimo pagado es: {min_totalpaybenefits:.2f}")

#Averiguamos cuantas veces se repite ese valor minimo
cantidad_min_totalpaybenefits = (salarios["TotalPayBenefits"] == min_totalpaybenefits).sum()
print(f"El valor minimo de TotalPayBenefits se repite {cantidad_min_totalpaybenefits} veces")

#Miramos quien es la persona que se gana ese dinero
print(f"El empleado que menos gana es: {salarios.loc[salarios["TotalPayBenefits"] == min_totalpaybenefits, "EmployeeName"].values[0]}")
```

**Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?**

crea una nueva columna donde se guarde la longitud del titulo del trabajo. tip: usar el método `.apply()`

```python
# Copie el código aca
#Creamos la nueva columns llamada longitud, que nos dice la longitud del nombre del titulo de trabajo de cada empleado
salarios["longitud"] = salarios["JobTitle"].apply(lambda x: len(str(x)))

salarios.head(5)
```

```python
# Correlación entre Longitud del titulo de trabajo y el salario total
correlacion = salarios["longitud"].corr(salarios["TotalPayBenefits"])
print(f"La correlación entre la longitud del titulo de trabajo y el salario total es: {correlacion:.4f}")
```


*Al ser una correlación negativa (-0.0369) quiere decir que no existe una relación entre la longitud del titulo de trabajo y el salario total.*
