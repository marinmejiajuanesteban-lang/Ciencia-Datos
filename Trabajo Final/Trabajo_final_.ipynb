```python
from scipy.stats import kstest

for col in cols:
    data = futuro_IA[col]

    # estandarizar
    data_std = (data - np.mean(data)) / np.std(data)

    stat, p = kstest(data_std, 'norm')

    print(f"\nVariable: {col}")
    print(f"Estadístico: {stat:.4f}")
    print(f"p-valor: {p:.4f}")

    if p > 0.05:
        print("→ Aproximadamente normal")
    else:
        print("→ No sigue distribución normal")
```

Se realizaron pruebas de normalidad utilizando el test de Kolmogorov-Smirnov para evaluar la distribución de las variables numéricas.

Los resultados indican que todas las variables presentan p-valores inferiores a 0.05, lo que lleva a rechazar la hipótesis nula de normalidad. Esto implica que ninguna de las variables sigue una distribución normal.

Estos resultados son consistentes con el análisis visual previo mediante histogramas, donde se identificaron distribuciones sesgadas, uniformes o discretas, particularmente en variables como el salario, las vacantes y la demanda de habilidades.

Es importante destacar que la variable de estabilidad laboral, al ser de tipo discreto, no es adecuada para este tipo de pruebas, lo que refuerza la necesidad de seleccionar correctamente las técnicas estadísticas según la naturaleza de los datos.

Dado que las variables no siguen una distribución normal, el análisis posterior deberá basarse en métodos no paramétricos, como la correlación de Spearman, en lugar de técnicas que asumen normalidad como la correlación de Pearson.

## Feature Engineering

###Variable: nivel_laboral

Codificamos nivel_experienci y nivel_educacion como variables ordinales y las sumamos para obtener un **índice de perfil profesional**.

| Experiencia | Valor | Educación | Valor |
|---|---|---|---|
| Entry | 1 | High School | 1 |
| Mid | 2 | Bachelor | 2 |
| Senior | 3 | Master | 3 |
| — | — | PhD | 4 |

`nivel_laboral = exp_num + edu_num` → rango de **3 a 7**

```python
# Mapeos ordinales
map_exp = {"Entry": 1, "Mid": 2, "Senior": 3}
map_edu = {"High School": 1, "Bachelor": 2, "Master": 3, "PhD": 4}

futuro_IA["exp_num"] = futuro_IA["nivel_experiencia"].map(map_exp)
futuro_IA["edu_num"] = futuro_IA["nivel_educacion"].map(map_edu)
futuro_IA["nivel_laboral"] = futuro_IA["exp_num"] + futuro_IA["edu_num"]

print("Distribución de nivel_laboral:")
display(futuro_IA["nivel_laboral"].value_counts().sort_index().to_frame("conteo"))

print(f"\nRango: {futuro_IA['nivel_laboral'].min()} – {futuro_IA['nivel_laboral'].max()}")
print(f"Media: {futuro_IA['nivel_laboral'].mean():.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Distribución de nivel_laboral
sns.countplot(data=futuro_IA, x="nivel_laboral", hue="nivel_laboral", palette="Blues_d", legend=False, ax=axes[0])
axes[0].set_title("Distribución de nivel_laboral")
axes[0].set_xlabel("Puntuación (exp_num + edu_num)")
axes[0].set_ylabel("Cantidad de empleos")

# nivel_laboral vs estabilidad laboral
sns.boxplot(data=futuro_IA, x="estabilidad_laboral", y="nivel_laboral", hue="estabilidad_laboral", palette="Set2", legend=False, ax=axes[1])
axes[1].set_title("nivel_laboral por Estabilidad Laboral")
axes[1].set_xlabel("Estabilidad laboral (0=Riesgo, 1=Estable, 2=Crecimiento)")

plt.suptitle("Feature Engineering: nivel_laboral", fontweight="bold", fontsize=13)
plt.tight_layout()
plt.show()
```

La variable nivel_laboral, construida a partir de la combinación del nivel de experiencia y educación, presenta una mayor concentración en valores intermedios, especialmente alrededor de la puntuación 5, lo que indica que predominan perfiles con un equilibrio entre formación y experiencia. Los niveles más bajos y más altos tienen menor representación, sugiriendo que hay menos casos de perfiles muy junior o altamente especializados.

En relación con la estabilidad laboral, se observa una tendencia creciente del nivel_laboral a medida que mejora la estabilidad (de riesgo a crecimiento), lo que indica que los perfiles con mayor puntuación tienden a estar asociados con empleos más estables o con mejores perspectivas. Sin embargo, también se evidencia cierta dispersión dentro de cada categoría, lo que sugiere que la estabilidad laboral no depende exclusivamente del nivel laboral, sino que puede estar influenciada por otros factores adicionales

```python
# Ratio de eficiencia
futuro_IA["ratio_eficiencia"] = futuro_IA["salario_usd"] / futuro_IA["vacantes"].replace(0, np.nan)

print("Estadísticas del ratio de eficiencia:")
display(futuro_IA["ratio_eficiencia"].describe().to_frame())
```

El ratio_eficiencia indica cuántos USD de salario corresponden a cada vacante disponible. Un ratio alto significa que hay pocas vacantes pero bien remuneradas (roles especializados y escasos), mientras que un ratio bajo implica muchas vacantes con salarios relativamente bajos (roles masivos y menos exclusivos).

```python
# Distribución del ratio de eficiencia
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

sns.countplot(data=futuro_IA, x="nivel_laboral", hue="nivel_laboral", palette="Blues_d", legend=False, ax=axes[0])
axes[0].set_title("Distribución de nivel_laboral")
axes[0].set_xlabel("Puntuación (exp_num + edu_num)")
axes[0].set_ylabel("Cantidad de empleos")

sns.boxplot(data=futuro_IA, x="estabilidad_laboral", y="nivel_laboral", hue="estabilidad_laboral", palette="Set2", legend=False, ax=axes[1])
axes[1].set_title("nivel_laboral por Estabilidad Laboral")
axes[1].set_xlabel("Estabilidad laboral (0=Riesgo, 1=Estable, 2=Crecimiento)")

plt.tight_layout()
plt.show()
```

Un ratio de eficiencia más alto en empleos con bajo riesgo de IA sugiere que los trabajos más exclusivos y bien remunerados (con pocas vacantes) tienden a estar menos expuestos a la automatización, mientras que roles masivos y menos remunerados son más vulnerables.

```python
# Normalizar probabilidad_riesgo_ia a rango [0, 1]
riesgo_min = futuro_IA["probabilidad_riesgo_ia"].min()
riesgo_max = futuro_IA["probabilidad_riesgo_ia"].max()
futuro_IA["riesgo_norm"] = (futuro_IA["probabilidad_riesgo_ia"] - riesgo_min) / (riesgo_max - riesgo_min)

# Normalizar nivel_laboral a rango [0, 1]
nivel_min = futuro_IA["nivel_laboral"].min()
nivel_max = futuro_IA["nivel_laboral"].max()
futuro_IA["nivel_norm"] = (futuro_IA["nivel_laboral"] - nivel_min) / (nivel_max - nivel_min)

# Índice de vulnerabilidad:
# Alto riesgo de IA (riesgo_norm alto) + Bajo nivel laboral (1 - nivel_norm alto)
# Se pondera con 60% al riesgo de IA y 40% al nivel laboral
futuro_IA["indice_vulnerabilidad"] = 0.6 * futuro_IA["riesgo_norm"] + 0.4 * (1 - futuro_IA["nivel_norm"])

print("Estadísticas del índice de vulnerabilidad:")
display(futuro_IA["indice_vulnerabilidad"].describe().to_frame())
```

El indice_vulnerabilidad oscila entre 0 y 1. Un valor cercano a 1 indica alta vulnerabilidad (mucho riesgo de ser reemplazado por IA + perfil laboral débil), mientras que un valor cercano a 0 indica baja vulnerabilidad (poco riesgo + perfil laboral fuerte)

**Escalado y normalización**

En el dataset hay variables como salario_usd que tienen valores muy grandes, mientras que otras como probabilidad_riesgo_ia tienen valores pequeños. Si no se escalan, el modelo puede darle más importancia al salario solo por su tamaño.

Para solucionar esto, se aplicó un proceso de escalado.

Método elegido: RobustScaler

Ya que este método usa la mediana y no se ve tan afectado por estos valores atípicos.

¿Por qué no se usaron otros métodos?

StandardScaler: usa la media, que se puede afectar mucho por outliers
MinMaxScaler: depende del valor mínimo y máximo, por lo que los valores extremos pueden distorsionar el resultado
MaxAbsScaler: se usa más en datos con muchos ceros (como texto), no en este caso
Normalizer: trabaja por filas y no por columnas, por lo que no es adecuado para este problema

Finalmente, las variables binarias (como las creadas con OneHotEncoding) y las variables ordinales no necesitan escalado, ya que ya están en rangos pequeños. Sin embargo, incluirlas no afecta significativamente el modelo.

## Definición de Features y Variable Objetivo

Antes de aplicar PCA necesitamos definir claramente:
- **X**: las variables predictoras (numéricas escaladas + OHE)
- **y**: la variable objetivo `nivel_riesgo_ia_enc` (ya codificada como 0, 1, 2)

Se excluyen variables que generarían fuga de información (*data leakage*):
- `probabilidad_riesgo_ia` → es la fuente directa del target
- `indice_vulnerabilidad` y `riesgo_norm` → construidas con el target
- `nivel_riesgo_ia` y `nivel_riesgo_ia_enc` → son el target mismo

```python
# Variable objetivo — mapeo directo desde texto
mapa_riesgo   = {'Low Risk': -1, 'Medium Risk': 0, 'High Risk': 1}
mapa_etiquetas = {-1: 'Low Risk', 0: 'Medium Risk', 1: 'High Risk'}

y = futuro_IA['nivel_riesgo_ia'].map(mapa_riesgo).astype(int)

print(f'Shape de y: {y.shape}')
print(f'\nDistribución de la variable objetivo:')
print(y.map(mapa_etiquetas).value_counts())
```

Aqui dividimos el dataset **antes** de aplicar PCA para evitar *data leakage*.
Si ajustáramos PCA sobre todos los datos (incluido el test), estaríamos
filtrando información del conjunto de evaluación al modelo.

- **80% entrenamiento** → para ajustar PCA y entrenar el modelo
- **20% prueba** → para evaluar el desempeño real
- `stratify=y` garantiza que las proporciones de cada clase se mantengan en ambos conjuntos

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f'Train: {X_train.shape[0]} filas  |  Test: {X_test.shape[0]} filas')
print(f'\nProporción de clases en Train:')
print((y_train.map(mapa_etiquetas).value_counts(normalize=True) * 100).round(1).astype(str) + '%')
print(f'\nProporción de clases en Test:')
print((y_test.map(mapa_etiquetas).value_counts(normalize=True) * 100).round(1).astype(str) + '%')
```

El PCA se ajusta **solo sobre X_train** y luego se aplica (transforma) tanto al train como al test.

Para elegir el número de componentes usamos dos criterios visuales:
1. **Scree Plot**: buscar el "codo" donde la varianza explicada cae bruscamente
2. **Varianza acumulada**: elegir el mínimo de componentes que retienen ≥ 95% de la varianza

```python
pca_exploratorio = PCA(random_state=42)
pca_exploratorio.fit(X_train)

varianza_exp  = pca_exploratorio.explained_variance_ratio_
varianza_acum = np.cumsum(varianza_exp)
n_comp_90     = int(np.argmax(varianza_acum >= 0.90) + 1)
n_comp_95     = int(np.argmax(varianza_acum >= 0.95) + 1)
n_comp_99     = int(np.argmax(varianza_acum >= 0.99) + 1)

print('Varianza explicada por componente:')
for i, (v, acum) in enumerate(zip(varianza_exp, varianza_acum)):
    marca = ''
    if i + 1 == n_comp_90: marca = '  ← 90%'
    if i + 1 == n_comp_95: marca = '  ← 95%'
    if i + 1 == n_comp_99: marca = '  ← 99%'
    print(f'  PC{i+1:02d}: {v:.4f}  |  Acumulada: {acum:.4f}{marca}')

print(f'\nResumen:')
print(f'  {n_comp_90} componentes → 90% de varianza')
print(f'  {n_comp_95} componentes → 95% de varianza')
print(f'  {n_comp_99} componentes → 99% de varianza')
```

```python
n = len(varianza_exp)
x = range(1, n + 1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scree Plot
axes[0].bar(x, varianza_exp, color='steelblue', alpha=0.7)
axes[0].plot(x, varianza_exp, 'o-', color='darkblue', markersize=4)
axes[0].set_title('Scree Plot — Varianza por Componente')
axes[0].set_xlabel('Componente Principal')
axes[0].set_ylabel('Varianza Explicada')
axes[0].set_xticks(list(x))

# Varianza Acumulada
axes[1].plot(x, varianza_acum, 'ro-', markersize=4)
for umbral, color, label in [(0.90, 'green', '90%'), (0.95, 'orange', '95%'), (0.99, 'red', '99%')]:
    n_c = int(np.argmax(varianza_acum >= umbral) + 1)
    axes[1].axhline(y=umbral, color=color, linestyle='--', alpha=0.7, label=f'{label} → PC{n_c}')
    axes[1].axvline(x=n_c, color=color, linestyle=':', alpha=0.5)
axes[1].set_title('Varianza Acumulada')
axes[1].set_xlabel('Número de Componentes')
axes[1].set_ylabel('Varianza Acumulada')
axes[1].legend()
axes[1].set_xticks(list(x))

fig.suptitle('Análisis PCA — Selección del Número de Componentes',
             fontsize=14, fontweight='bold', color='darkblue')
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.show()
```

**Decisión:** Se seleccionan los componentes necesarios para retener el **95% de la varianza**.

Este umbral equilibra reducción de dimensionalidad con preservación de información.
Retener menos (90%) puede sacrificar información importante; retener más (99%) puede
incluir ruido que perjudica la generalización del modelo.

```python
pca = PCA(n_components=n_comp_95, random_state=42)

# fit SOLO en train → transform en ambos
X_train_pca = pca.fit_transform(X_train)
X_test_pca  = pca.transform(X_test)

varianza_retenida = pca.explained_variance_ratio_.sum()

print(f'PCA aplicado:')
print(f'  Dimensiones originales : {X_train.shape[1]} variables')
print(f'  Dimensiones reducidas  : {X_train_pca.shape[1]} componentes')
print(f'  Varianza retenida      : {varianza_retenida:.4f} ({varianza_retenida*100:.1f}%)')
print(f'\nShape X_train_pca: {X_train_pca.shape}')
print(f'Shape X_test_pca : {X_test_pca.shape}')
```

## Curvas ROC y AUC

**¿Qué es la curva ROC?**
ROC (Receiver Operating Characteristic) es una curva que muestra la capacidad
del modelo para distinguir entre clases. Para cada clase se grafica:
- **Eje X (False Positive Rate)**: qué tan seguido el modelo se equivoca
  prediciendo esa clase cuando no lo es
- **Eje Y (True Positive Rate / Recall)**: qué tan bien detecta los casos
  reales de esa clase

Una curva que se acerca a la esquina superior izquierda indica un modelo
que distingue muy bien esa clase. La línea diagonal punteada representa
un clasificador aleatorio (sin ninguna capacidad predictiva).

**¿Qué es el AUC?**
AUC (Area Under the Curve) es el área bajo la curva ROC. Resume en un solo
número qué tan bien separa el modelo cada clase:
- **AUC = 1.0** → separación perfecta
- **AUC = 0.5** → equivale a adivinar al azar
- **AUC > 0.85** → se considera un modelo con buen desempeño

**¿Para qué lo usamos aquí?**
Como tenemos 3 clases, se usa el enfoque **One-vs-Rest (OvR)**: se evalúa
cada clase contra las otras dos de forma independiente. Esto permite identificar
cuáles clases el modelo distingue mejor y cuáles le generan más dificultad,
complementando lo visto en la matriz de confusión.

```python
auc_score = roc_auc_score(y_test, y_proba, multi_class='ovr', average='macro')
print(f'ROC-AUC Macro (OvR): {auc_score:.4f}')

fig, ax = plt.subplots(figsize=(8, 6))

colores_roc = ['#2196F3', '#FF9800', '#F44336']
clases      = [-1, 0, 1]
nombres     = ['Low Risk', 'Medium Risk', 'High Risk']

for i, (clase, nombre, color) in enumerate(zip(clases, nombres, colores_roc)):
    RocCurveDisplay.from_predictions(
        (y_test == clase).astype(int),
        y_proba[:, i],
        name=nombre,
        color=color,
        ax=ax
    )

ax.plot([0, 1], [0, 1], 'k--', label='Clasificador aleatorio')
ax.set_title(f'Curvas ROC por Clase (OvR)\nAUC Macro = {auc_score:.3f}',
             fontsize=12, fontweight='bold')
ax.legend(loc='lower right')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Interpretación de las Curvas ROC:**

El **AUC Macro de 0.899** indica que en promedio el modelo tiene una excelente
capacidad de separación entre clases. Sin embargo, el AUC no mide cuántos casos
predice bien en total — eso lo mide la accuracy (74%). Lo que el AUC dice es
**qué tan bien el modelo distingue cada clase de las demás**.

Analizando por clase:

- **High Risk (AUC=0.99)**: separación casi perfecta. La curva roja sube
  verticalmente hacia la esquina superior izquierda, lo que confirma que el

- **Low Risk (AUC=0.87)**: buena separación. El modelo distingue bien los

- **Medium Risk (AUC=0.84)**: es la clase más difícil de separar, consistente

En resumen: el modelo no predice correctamente el 89% de los casos individuales,

## Validación Cruzada Estratificada

La validación cruzada permite estimar el rendimiento real del modelo de forma
más robusta que una sola división train/test. Una única división puede dar
resultados optimistas o pesimistas dependiendo de cómo quedaron los datos.

Se usa **StratifiedKFold con 5 folds**: el dataset se divide en 5 partes,
el modelo se entrena 5 veces usando 4 partes y evaluando en la restante.
`stratify` garantiza que cada fold conserve las proporciones de las 3 clases.

Si las métricas de la validación cruzada son similares a las del test,
confirma que el modelo **generaliza bien** y no depende de un split particular.

```python
X_pca_completo = pca.transform(X)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores_acc = cross_val_score(modelo, X_pca_completo, y, cv=cv, scoring='accuracy')
scores_f1  = cross_val_score(modelo, X_pca_completo, y, cv=cv, scoring='f1_macro')
scores_auc = cross_val_score(modelo, X_pca_completo, y, cv=cv, scoring='roc_auc_ovr_weighted')

print('=' * 55)
print('VALIDACIÓN CRUZADA — 5 Folds Estratificados')
print('=' * 55)
print(f'Accuracy   : {scores_acc.mean():.4f} ± {scores_acc.std():.4f}')
print(f'F1 Macro   : {scores_f1.mean():.4f} ± {scores_f1.std():.4f}')
print(f'ROC-AUC OvR: {scores_auc.mean():.4f} ± {scores_auc.std():.4f}')
print()
print('Detalle por fold:')
for i, (acc, f1, auc) in enumerate(zip(scores_acc, scores_f1, scores_auc), 1):
    print(f'  Fold {i}: Acc={acc:.4f}  F1={f1:.4f}  AUC={auc:.4f}')
```

```python
metricas = {'Accuracy': scores_acc, 'F1 Macro': scores_f1, 'ROC-AUC': scores_auc}

fig, ax = plt.subplots(figsize=(10, 5))

x_pos = np.arange(5)
width = 0.25
colores_met = ['steelblue', 'darkorange', 'seagreen']

for i, (nombre, scores) in enumerate(metricas.items()):
    ax.bar(x_pos + i * width, scores, width,
           label=f'{nombre} (μ={scores.mean():.3f})',
           color=colores_met[i], alpha=0.8)
    ax.axhline(scores.mean(), color=colores_met[i],
               linestyle='--', alpha=0.5, linewidth=1)

ax.set_xticks(x_pos + width)
ax.set_xticklabels([f'Fold {i}' for i in range(1, 6)])
ax.set_ylim(0, 1.05)
ax.set_ylabel('Puntuación')
ax.set_title('Métricas por Fold — Validación Cruzada Estratificada',
             fontsize=12, fontweight='bold')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Interpretación de la Validación Cruzada:**

Los resultados de la validación cruzada confirman que el modelo es **estable
y generaliza bien**. Tres observaciones clave:

**1. Consistencia entre folds:**
Las métricas varían muy poco entre los 5 folds. La desviación estándar es
mínima en las tres métricas (±0.01 en Accuracy, ±0.008 en F1, ±0.005 en AUC),
lo que indica que el modelo no depende de cómo se dividan los datos.

**2. Coherencia con los resultados del test:**
- Accuracy CV (73.1%) vs Accuracy Test (74%) → prácticamente iguales
- F1 Macro CV (76.5%) vs F1 Macro Test (77%) → prácticamente iguales
- AUC CV (87.3%) vs AUC Test (89.9%) → muy cercanos

Esta coherencia confirma que **no hay sobreajuste (overfitting)**. El modelo
aprendió patrones generales del mercado laboral y no simplemente memorizó
los datos de entrenamiento.

**3. Fold 1 ligeramente superior:**
El primer fold muestra métricas un poco más altas que los demás, pero la
diferencia es mínima y dentro del rango normal de variabilidad estadística.

En conclusión, el pipeline PCA + Regresión Logística produce un modelo
robusto, estable y con buen desempeño para predecir el nivel de riesgo
de automatización por IA en el mercado laboral.
